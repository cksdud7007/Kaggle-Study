```python
%matplotlib inline

# backbone
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns

sns.set(style = 'ticks', context = 'talk', font_scale= 1)
plt.style.use("dark_background")
```

```python
train = pd.read_csv('C:\\Users\\이찬영\\Desktop\\House_Price\\train.csv')
test = pd.read_csv('C:\\Users\\이찬영\\Desktop\\House_Price\\test.csv')
data = pd.concat([train, test], axis = 0)
```

## Interaction check

```python
sns.lmplot(
    x='GrLivArea', y="SalePrice", hue="OverallQual", col="OverallQual",
    data=data, scatter_kws={"edgecolor": 'w','s':30}, col_wrap=5, height=4,
)
```
